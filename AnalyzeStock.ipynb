```python
import requests
from bs4 import BeautifulSoup
```

```python
tesla_url="https://finance.yahoo.com/quote/TSLA/history"
headers = {'User-Agent': 'Mozilla/5.0'}
```

```python
tesla_data=requests.get(tesla_url,headers=headers).text
```

```python
tesla_data
```

```python
soup=BeautifulSoup(tesla_data,'html.parser')
```

```python
tsla_data=pd.DataFrame(columns=["Date","Open","High","Low","Close","Volume"])
```

```python
for row in soup.find('tbody').find_all('tr'):
    col=row.find_all('td')
    date=col[0].text
    Open=col[1].text
    High=col[2].text
    Low=col[3].text
    Close=col[4].text
    Adj_Close=col[5].text
    volume=col[6].text
```

```python
tsla_data=pd.concat([tsla_data,pd.DataFrame({"Date":[date],"Open":[Open],"High":[High],"Low":[Low],"Close":[Close],"Volume":[volume]})],ignore_index=True)
```

```python
tsla_data.head()
```

```python
import matplotlib.pyplot as plt

def make_graph(data, title="Tesla Stock Price"):
    plt.figure(figsize=(12, 6))
    plt.plot(data.index, data['Close'], label='Close Price', color='blue')
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.title(title)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()
    
make_graph(tesla_history, title="Tesla Stock Closing Prices - Last 1 Month")
```
